```python
import pandas as pd
import calendar
import os
wkdays = []
for i in range(7):
    wkdays.append(calendar.day_abbr[i])
result = {}

flist = []
for f in os.listdir("PARKING_LOTS"):
    if f.endswith(".csv"): flist.append(f)

for f in flist:
    df = pd.read_csv("PARKING_LOTS/"+f,encoding = 'unicode_escape')
    avglist = {}
    for day in wkdays:
        df1 = df.query("weekday==@day")
        avglist[day] = df1["occupacy_percentage"].mean()
    result[f] = avglist
result
```
